# lists

```python
import numpy as np
aliza =[2,4,5,8,96,5,4,]
print("original list",aliza)
print("decending",sorted(aliza, reverse=True))
print("ascending",sorted(aliza))
print("sum",sum(aliza))
print("max",max(aliza))
print("min",min(aliza))
aliza.append(67)
aliza.remove(8)
print("updated list",aliza)


```

# tuple

```python
t = (10, 20, 30)
print("1st element",t[0])
print("last element",t[-1])
```

# sets

```python
my_set = {1, 2, 3, 4, 4, 5}
set2 = {4, 5, 6, 7}
print("Set:", my_set)
print("Union:", my_set.union(set2))
print("Intersection:", my_set.intersection(set2))
print("Difference:", my_set.difference(set2))
```

# dictionary

```python
students = {
    "Ali": 85,
    "Sara": 90,
    "Ahmed": 78,
    "Ayesha": 88,
    "Zain": 92
}
print("keys:", students.keys())
print("values:", students.values())
```
