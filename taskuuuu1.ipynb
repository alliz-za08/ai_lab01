```python
import heapq  # for priority queue

def ucs(graph, start, goal):
    pq = []
    heapq.heappush(pq, (0, start))  # push start node with cost 0

    visited = set()  # to keep track of visited nodes

    while pq:
        cost, node = heapq.heappop(pq)  # get node with lowest cost

        if node in visited:
            continue  # skip if already visited

        visited.add(node)

        print("Visited:", node, "Cost:", cost)

        if node == goal:
            print("Goal reached with cost:", cost)
            return  # stop when goal is found

        for neighbor, weight in graph.get(node, []):
            # add neighbors with updated cost
            heapq.heappush(pq, (cost + weight, neighbor))


# Graph representation
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 8), ('G', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)]
}

ucs(graph, 'S', 'G')
```
