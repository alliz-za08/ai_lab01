```python
import random

# Example function to maximize
# f(x) = -(x^2) + 10  (maximum at x = 0)
def objective_function(x):
    return -(x ** 2) + 10

# Generate neighbors (small changes)
def get_neighbors(x):
    return [x - 1, x + 1]

def hill_climbing(start):
    current = start

    while True:
        neighbors = get_neighbors(current)
        next_node = max(neighbors, key=objective_function)

        # If no better neighbor, stop
        if objective_function(next_node) <= objective_function(current):
            break

        current = next_node

    return current

# Run the algorithm
start = random.randint(-10, 10)
result = hill_climbing(start)

print("Start:", start)
print("Best solution found:", result)
print("Best value:", objective_function(result))
```
