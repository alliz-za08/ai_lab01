# Activity one

```python


link_data = {
    'N1': ['N2', 'N3'],
    'N2': ['N4', 'N5'],
    'N3': ['N6'],
    'N4': [],
    'N5': ['N6'],
    'N6': []
}

seen_list = []
waiting_line = []

def walk_bfs(memory, structure, start_key):
    memory.append(start_key)
    waiting_line.append(start_key)

    while waiting_line:
        element = waiting_line.pop(0)
        print(element, end=" ")

        for item in structure[element]:
            if item not in memory:
                memory.append(item)
                waiting_line.append(item)

print("BFS Output:")
walk_bfs(seen_list, link_data, 'N1')
```

# Activity 2

```python

def display_parents(current_node, tree_data, prev):
    if prev == 0:
        print(current_node, "-> MainRoot")
    else:
        print(current_node, "->", prev)

    for nxt in tree_data[current_node]:
        if nxt != prev:
            display_parents(nxt, tree_data, current_node)

def display_children(start_root, tree_data):
    process_line = []
    mark_visit = [0] * len(tree_data)

    process_line.append(start_root)

    while process_line:
        now = process_line.pop(0)
        mark_visit[now] = 1

        print(now, "-> ", end=" ")

        for nxt in tree_data[now]:
            if mark_visit[nxt] == 0:
                print(nxt, end=" ")
                process_line.append(nxt)
        print()


# Show leaf nodes
def display_leaves(start_root, tree_data):
    for idx in range(len(tree_data)):
        if len(tree_data[idx]) == 1 and idx != start_root:
            print(idx, end=" ")
    print()

def display_degree(start_root, tree_data):
    for idx in range(1, len(tree_data)):
        print(idx, ": ", end="")

        if idx == start_root:
            print(len(tree_data[idx]))
        else:
            print(len(tree_data[idx]) - 1)

count_nodes = 7
root_node = 1

tree_box = [[] for _ in range(count_nodes + 1)]

tree_box[1].append(2)
tree_box[2].append(1)

tree_box[1].append(3)
tree_box[3].append(1)

tree_box[1].append(4)
tree_box[4].append(1)

tree_box[2].append(5)
tree_box[5].append(2)

tree_box[2].append(6)
tree_box[6].append(2)

tree_box[4].append(7)
tree_box[7].append(4)


print("Parent List:")
display_parents(root_node, tree_box, 0)

print("\nChildren List:")
display_children(root_node, tree_box)

print("\nLeaf List:")
display_leaves(root_node, tree_box)

print("Degree List:")
display_degree(root_node, tree_box)
```

# Activity 3

```python

path_map = {
    'S': ['T', 'U'],
    'T': ['V', 'W'],
    'U': ['X'],
    'V': [],
    'W': ['X'],
    'X': []
}

checked = []
buffer_line = []

def start_traversal(data_map, origin):
    checked.append(origin)
    buffer_line.append(origin)

    while buffer_line:
        current_node = buffer_line.pop(0)
        print(current_node, end=" ")

        for nxt in data_map[current_node]:
            if nxt not in checked:
                checked.append(nxt)
                buffer_line.append(nxt)

print("Activity 3 BFS:")
start_traversal(path_map, 'S')
```
