```python

class FigureTrap:
    def __init__(self, side_a, side_b, height_a):
        self.side_a = side_a
        self.side_b = side_b
        self.height_a = height_a

    def area_value(self):
        return 0.5 * (self.side_a + self.side_b) * self.height_a

class FigurePara:
    def __init__(self, base_a, height_b):
        self.base_a = base_a
        self.height_b = height_b

    def area_value(self):
        return self.base_a * self.height_b

data_pack = [
    (4, 6, 5, 7, 3),
    (8, 10, 4, 5, 6),
    (3, 9, 2, 6, 4)
]

counter = 1

for item in data_pack:
    n1, n2, n3, n4, n5 = item

    shape_one = FigureTrap(n1, n2, n3)
    shape_two = FigurePara(n4, n5)

    result_box = {
        "Trapezoid": shape_one.area_value(),
        "Parallelogram": shape_two.area_value()
    }

    print("\nSet", counter)
    print("Areas:", result_box)

    best_shape = max(result_box, key=result_box.get)

    print("Maximum Area Shape:", best_shape)
    print("Area:", result_box[best_shape])

    counter += 1
```
