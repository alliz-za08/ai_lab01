```python
# ============================================================
# Snake Game - Lab Task 2 (Student Task)
# Advanced Features Implementation
# Run in Google Colab - Just Shift+Enter!
# ============================================================

from IPython.display import display, HTML

html_code = """
<html>
<head>
<style>
  body { background:#0f0e17; display:flex; flex-direction:column;
         align-items:center; font-family:Arial; color:white; margin:0; padding:10px; }
  h2   { color:#ff6b35; margin:5px 0; letter-spacing:2px; font-size:22px; }
  #info{ display:flex; gap:40px; margin:8px 0; font-size:16px; font-weight:bold;
         background:#1a1a2e; padding:8px 30px; border-radius:20px; }
  #scoreBox  { color:#00ff88; }
  #levelBox  { color:#ff6b35; }
  #speedBox  { color:#ffaa00; }
  #tipBox    { color:#ffee00; font-size:13px; min-height:18px; margin-bottom:4px; }
  canvas     { border:3px solid #ff6b35; border-radius:4px; background:#000; display:block; }
  #msg       { margin-top:10px; font-size:18px; color:#ff6b35; min-height:24px; text-align:center; }
  #restartBtn{ display:none; margin-top:8px; padding:10px 35px;
               background:#ff6b35; color:white; border:none; border-radius:8px;
               font-size:16px; cursor:pointer; font-weight:bold; }
  #restartBtn:hover { background:#e55a2b; }
  #controls  { margin-top:6px; color:#888; font-size:13px; }
  #features  { margin-top:8px; background:#1a1a2e; border-radius:10px;
               padding:8px 20px; font-size:13px; color:#aaa; text-align:center; }
  #features span { margin:0 10px; }
  .feat-on   { color:#00ff88 !important; font-weight:bold; }
</style>
</head>
<body>
<h2>🐍 SNAKE GAME — LAB TASK 2</h2>

<div id="info">
  <span id="scoreBox">Score: 0</span>
  <span id="levelBox">Level: 1</span>
  <span id="speedBox">Speed: Normal</span>
</div>

<div id="tipBox" id="tipBox"></div>

<canvas id="c" width="600" height="600" tabindex="0"></canvas>

<!-- Feature Status Bar -->
<div id="features">
  <span id="f1">⚡ Speed Boost (score&gt;100): OFF</span> |
  <span id="f2">⭐ Yellow Fruit (score&gt;150): OFF</span> |
  <span id="f3">🧱 Obstacles: ON</span>
</div>

<div id="msg"></div>
<button id="restartBtn" onclick="startGame()">▶ Restart</button>
<div id="controls">⬆⬇⬅➡ Arrow Keys  or  W A S D</div>

<script>
const canvas = document.getElementById('c');
const ctx    = canvas.getContext('2d');

const CELL     = 20;
const COLS     = 30;
const ROWS     = 30;
const BASE_FPS = 8;

// ── TASK 2.3: Obstacles / inner walls ──────────────────────────────
// Three different wall shapes inside the game
const WALLS = new Set([
  // Horizontal wall top-left area
  '5,5','6,5','7,5','8,5','9,5',
  // Vertical wall right area
  '22,8','22,9','22,10','22,11','22,12',
  // L-shaped wall bottom-center
  '12,22','13,22','14,22','15,22',
  '12,23','12,24',
  // Small block top-right
  '24,4','25,4','24,5','25,5',
]);

let snake, snakeSet, dir, nextDir;
let score, rFruit, yFruit, fps, level;
let gameRunning, loopId;
let speedBoosted = false;
let yellowUnlocked = false;

function key(x,y){ return x+','+y; }

// ── Fast free cell (no list scanning) ─────────────────────────────
function freeCell(ex1, ex2){
  let p;
  do {
    p = {
      x: Math.floor(Math.random()*(COLS-2))+1,
      y: Math.floor(Math.random()*(ROWS-2))+1
    };
  } while(
    snakeSet.has(key(p.x,p.y)) ||
    WALLS.has(key(p.x,p.y))    ||
    (ex1 && p.x===ex1.x && p.y===ex1.y) ||
    (ex2 && p.x===ex2.x && p.y===ex2.y)
  );
  return p;
}

// ── Draw one cell ──────────────────────────────────────────────────
function drawCell(x, y, fill, inner){
  ctx.fillStyle = fill;
  ctx.fillRect(x*CELL, y*CELL, CELL, CELL);
  if(inner){
    ctx.fillStyle = inner;
    ctx.fillRect(x*CELL+3, y*CELL+3, CELL-6, CELL-6);
  }
}

// ── Draw glowing circle on fruit ──────────────────────────────────
function drawFruitGlow(x, y, color){
  ctx.fillStyle = color;
  ctx.beginPath();
  ctx.arc(x*CELL+10, y*CELL+10, 5, 0, Math.PI*2);
  ctx.fill();
}

// ── Full render ───────────────────────────────────────────────────
function drawGame(){
  ctx.fillStyle = '#000';
  ctx.fillRect(0,0,600,600);

  // Border (blue walls)
  for(let x=0;x<COLS;x++){
    drawCell(x,0,'#1e40af','#3b82f6');
    drawCell(x,ROWS-1,'#1e40af','#3b82f6');
  }
  for(let y=1;y<ROWS-1;y++){
    drawCell(0,y,'#1e40af','#3b82f6');
    drawCell(COLS-1,y,'#1e40af','#3b82f6');
  }

  // ── TASK 2.3: Draw inner obstacle walls ──────────────────────────
  for(let w of WALLS){
    let [wx,wy] = w.split(',').map(Number);
    drawCell(wx, wy, '#78716c', '#57534e');
    // brick line detail
    ctx.strokeStyle = '#44403c';
    ctx.lineWidth = 1;
    ctx.strokeRect(wx*CELL+1, wy*CELL+1, CELL-2, CELL-2);
  }

  // Red fruit
  drawCell(rFruit.x, rFruit.y, '#991b1b', '#dc2626');
  drawFruitGlow(rFruit.x, rFruit.y, '#fca5a5');

  // ── TASK 2.2: Yellow fruit (double points) ────────────────────────
  if(yFruit){
    drawCell(yFruit.x, yFruit.y, '#a16207', '#facc15');
    drawFruitGlow(yFruit.x, yFruit.y, '#fef08a');
    // Star marker on yellow fruit
    ctx.fillStyle = '#fff';
    ctx.font = '11px Arial';
    ctx.textAlign = 'center';
    ctx.fillText('x2', yFruit.x*CELL+10, yFruit.y*CELL+14);
  }

  // Snake
  for(let i=0; i<snake.length; i++){
    let s = snake[i];
    if(i===0){
      // Head - darker with eyes
      drawCell(s.x, s.y, '#14532d', '#16a34a');
      // Eyes
      ctx.fillStyle = '#fff';
      let ex = s.x*CELL, ey = s.y*CELL;
      ctx.fillRect(ex+4,  ey+4, 4, 4);
      ctx.fillRect(ex+12, ey+4, 4, 4);
      ctx.fillStyle='#000';
      ctx.fillRect(ex+5,  ey+5, 2, 2);
      ctx.fillRect(ex+13, ey+5, 2, 2);
    } else {
      // Body - gradient effect by position
      let shade = Math.max(30, 180 - i*3);
      drawCell(s.x, s.y, '#15803d', `rgb(0,${shade},50)`);
    }
  }

  // Speed boost visual effect (flashing border when fast)
  if(speedBoosted){
    let t = Date.now() % 600;
    if(t < 300){
      ctx.strokeStyle = '#f97316';
      ctx.lineWidth = 4;
      ctx.strokeRect(2,2,596,596);
    }
  }
}

// ── HUD update ────────────────────────────────────────────────────
function updateHUD(){
  document.getElementById('scoreBox').textContent = 'Score: ' + score;
  document.getElementById('levelBox').textContent = 'Level: ' + level;

  // ── TASK 2.1: Speed status ────────────────────────────────────────
  if(speedBoosted){
    document.getElementById('speedBox').textContent = '⚡ SPEED x2';
    document.getElementById('speedBox').style.color = '#f97316';
    document.getElementById('f1').textContent = '⚡ Speed Boost (score>100): ON';
    document.getElementById('f1').className = 'feat-on';
  } else {
    document.getElementById('speedBox').textContent = 'Speed: Normal';
    document.getElementById('speedBox').style.color = '#ffaa00';
    document.getElementById('f1').textContent = '⚡ Speed Boost (score>100): OFF';
    document.getElementById('f1').className = '';
  }

  // ── TASK 2.2: Yellow fruit status ────────────────────────────────
  if(yellowUnlocked){
    document.getElementById('f2').textContent = '⭐ Yellow Fruit (score>150): ON';
    document.getElementById('f2').className = 'feat-on';
    document.getElementById('tipBox').textContent = '⭐ Yellow fruit gives x2 points (20 pts)!';
  } else if(score > 100){
    document.getElementById('tipBox').textContent = '🎯 ' + (150-score) + ' more points to unlock Yellow Fruit!';
  } else {
    document.getElementById('tipBox').textContent = '';
  }
}

// ── Game Over ─────────────────────────────────────────────────────
function gameOver(){
  gameRunning = false;
  clearTimeout(loopId);

  ctx.fillStyle = 'rgba(0,0,0,0.72)';
  ctx.fillRect(0,0,600,600);

  ctx.textAlign = 'center';
  ctx.fillStyle = '#ff6b35';
  ctx.font = 'bold 50px Arial';
  ctx.fillText('GAME OVER', 300, 230);

  ctx.fillStyle = '#fff';
  ctx.font = 'bold 28px Arial';
  ctx.fillText('Score: ' + score, 300, 285);

  ctx.fillStyle = '#00ff88';
  ctx.font = '20px Arial';
  ctx.fillText('Level Reached: ' + level, 300, 325);

  ctx.fillStyle = '#aaa';
  ctx.font = '16px Arial';
  ctx.fillText('Click Restart to play again', 300, 370);

  document.getElementById('msg').textContent = '';
  document.getElementById('restartBtn').style.display = 'inline-block';
}

// ── Main game step ────────────────────────────────────────────────
function gameStep(){
  if(!gameRunning) return;

  dir = nextDir;
  let hx = snake[0].x + dir.x;
  let hy = snake[0].y + dir.y;

  // Collision: border, self, obstacle
  if(hx<=0 || hx>=COLS-1 || hy<=0 || hy>=ROWS-1 ||
     snakeSet.has(key(hx,hy)) || WALLS.has(key(hx,hy))){
    gameOver();
    return;
  }

  snake.unshift({x:hx, y:hy});
  snakeSet.add(key(hx,hy));
  let ate = false;

  // Eat red fruit → +10 pts
  if(hx===rFruit.x && hy===rFruit.y){
    score += 10;
    rFruit = freeCell(yFruit, null);
    ate = true;
  }

  // ── TASK 2.2: Eat yellow fruit → +20 pts (double) ────────────────
  if(yFruit && hx===yFruit.x && hy===yFruit.y){
    score += 20;
    yFruit = freeCell(rFruit, null);
    ate = true;
  }

  if(!ate){
    let tail = snake.pop();
    snakeSet.delete(key(tail.x, tail.y));
  }

  // ── TASK 2.1: Double speed when score > 100 ───────────────────────
  if(score > 100 && !speedBoosted){
    speedBoosted = true;
    fps = BASE_FPS * 2;
  }

  // ── TASK 2.2: Unlock yellow fruit when score > 150 ────────────────
  if(score > 150 && !yellowUnlocked){
    yellowUnlocked = true;
    yFruit = freeCell(rFruit, null);
  }

  // Level system (every 50 pts)
  level = Math.floor(score / 50) + 1;

  drawGame();
  updateHUD();

  loopId = setTimeout(gameStep, 1000 / fps);
}

// ── Start / Restart ───────────────────────────────────────────────
function startGame(){
  clearTimeout(loopId);

  snake        = [{x:15, y:15}];
  snakeSet     = new Set([key(15,15)]);
  dir          = {x:1, y:0};
  nextDir      = {x:1, y:0};
  score        = 0;
  level        = 1;
  fps          = BASE_FPS;
  yFruit       = null;
  speedBoosted    = false;
  yellowUnlocked  = false;
  rFruit       = freeCell(null, null);
  gameRunning  = true;

  document.getElementById('msg').textContent = '';
  document.getElementById('restartBtn').style.display = 'none';
  document.getElementById('tipBox').textContent = '';
  document.getElementById('f1').className = '';
  document.getElementById('f2').className = '';
  document.getElementById('f1').textContent = '⚡ Speed Boost (score>100): OFF';
  document.getElementById('f2').textContent = '⭐ Yellow Fruit (score>150): OFF';

  updateHUD();
  gameStep();
}

// ── Keyboard controls ─────────────────────────────────────────────
document.addEventListener('keydown', function(e){
  const keys = ['ArrowUp','ArrowDown','ArrowLeft','ArrowRight','w','a','s','d','W','A','S','D'];
  if(keys.includes(e.key)) e.preventDefault();

  if((e.key==='ArrowUp'    || e.key==='w' || e.key==='W') && dir.y !==  1) nextDir={x:0,y:-1};
  if((e.key==='ArrowDown'  || e.key==='s' || e.key==='S') && dir.y !== -1) nextDir={x:0,y: 1};
  if((e.key==='ArrowLeft'  || e.key==='a' || e.key==='A') && dir.x !==  1) nextDir={x:-1,y:0};
  if((e.key==='ArrowRight' || e.key==='d' || e.key==='D') && dir.x !== -1) nextDir={x: 1,y:0};
});

// Auto start
startGame();
</script>
</body>
</html>
"""

display(HTML(html_code))
```
